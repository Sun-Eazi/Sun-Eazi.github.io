### 👁️ How Does an AI See? (Input -> Process -> Output)
We just hired a Global AI from Silicon Valley. Now, let's test it with a picture of a **Daladala** from the streets of Dar es Salaam.

To an AI, a picture is not a drawing; it is just a giant grid of numbers.

As Project Managers, we must understand the 3 logical steps of AI:

1. **INPUT:** We give the AI a local Tanzanian photo.

2. **PROCESS:** The mathematical brain calculates the probabilities inside the "Black Box."

3. **OUTPUT:** The AI translates its math into a human-readable table and draws bounding boxes.

**Action:** Run the code below and watch the AI process our Daladala!

````python
import os

# ==========================================
# STEP 1: INPUT (Provide the Data)
# ==========================================
# We fetch a real photo of a Daladala from the internet.
img_url = 'https://images.squarespace-cdn.com/content/v1/53baea7fe4b032d466a92a62/9c974051-2eea-47f0-b6ea-708e35727330/Dala+dala+3.jpg?format=2500w'
img_filename = 'daladala_test.jpg' # Changed the variable name slightly to match the context

# AI Engineers care about efficiency (and saving internet data!).
# We only download the image if we don't already have it.
if not os.path.exists(img_filename):
    try:
        print("📥 Downloading the image from the web...")
        torch.hub.download_url_to_file(img_url, img_filename)
        print(f"✅ Image download complete: {img_filename}")
    except Exception as e:
        print(f"❌ Image download failed: {e}")
else:
    print(f"✅ Image '{img_filename}' already exists! Skipping download.")

# ==========================================
# STEP 2: PROCESS (The AI Brain works)
# ==========================================
if os.path.exists(img_filename):
    print("\n🧠 The AI is looking at the image and calculating...")
    # This single line of code is where billions of math calculations happen!
    # The image goes into the 'model', and the answer comes out as 'results'.
    results = model(img_filename)

# ==========================================
# STEP 3: OUTPUT (Show us the evidence)
# ==========================================
    # 3-A: Visual Output (Draw bounding boxes on the image)
    print("🎨 Rendering the visual results...")
    results.show()

    # 3-B: Text Output (Print a data table)
    print("\n📊 [AI's Prediction Report]")
    # The raw result is too complex. We use 'pandas' to filter the data
    # and only show the 'name' (what it is) and 'confidence' (how sure it is).
    print(results.pandas().xyxy[0][['name', 'confidence']])
else:
    print("⚠️ No image to test. Please check the download step.")
````

## 🗣️ Phase 3: Teaching the AI Our Local Language (data.yaml)

Right now, our downloaded dataset is from Dhaka. The AI thinks a three-wheeler is called a "Rickshaw". But we are in Tanzania! We need to teach the AI our local language before we train it.

**Data Localization:** We will run a script to automatically translate the AI's dictionary (`data.yaml`):
* Rickshaw ➡️ **Bajaji**
* Motorbike ➡️ **Pikipiki**
* Bus/Van ➡️ **Daladala**

**Action:** You don't need to open the files and edit them manually. A Project Manager automates tasks! Just run the "Magic Translation Script" below.

````python
import yaml
import os

# ==========================================
# STEP 1: Open the AI's Dictionary (data.yaml)
# ==========================================
if 'base_dir' in globals() and os.path.exists(base_dir):
    target_yaml = os.path.join(base_dir, 'data.yaml')
    print(f"📖 Opening the AI's dictionary: {target_yaml}\n")

    # Load the existing YAML file
    with open(target_yaml, 'r') as f:
        data = yaml.safe_load(f)

# ==========================================
# STEP 2: Fix the GPS Map (Absolute Paths)
# ==========================================
    # We must lock the exact GPS coordinates so the AI doesn't get lost.
    # We change relative paths (like '../train') to Absolute paths.
    print("🗺️ Fixing the directory paths to Absolute Paths...")
    data['path'] = base_dir

    if 'train' in data:
        data['train'] = os.path.join(base_dir, 'train', 'images')
    if 'val' in data:
        data['val'] = os.path.join(base_dir, 'valid', 'images')
    if 'test' in data:
        data['test'] = os.path.join(base_dir, 'test', 'images')

# ==========================================
# STEP 3: The Magic Translation (Index Mapping)
# ==========================================
    # We map the specific Class IDs (0, 1, 2...) to Tanzanian terms.
    new_names = []

    for i, name in enumerate(data.get('names', [])):
        if i == 0:   new_names.append('Pikipiki (Motorcycle)')
        elif i == 1: new_names.append('Car')
        elif i == 2: new_names.append('Bajaj (Tricycle)')
        elif i == 3: new_names.append('Human (Pedestrian)')
        elif i == 4: new_names.append('Misc (Ignore)')
        elif i == 5: new_names.append('Daladala (Bus)')
        elif i == 6: new_names.append('Bicycle')
        elif i == 7: new_names.append('Rickshaw (Other)')
        else:        new_names.append(f'Unknown_{i}')

    # Apply the new Tanzanian words and update the class count
    data['names'] = new_names
    data['nc'] = len(new_names)

    print("\n🇹🇿 [Updated Tanzanian Class Dictionary]")
    for idx, final_name in enumerate(data['names']):
        print(f"  🟢 ID {idx}: {final_name}")

# ==========================================
# STEP 4: Save and Prepare for Training
# ==========================================
    # Write the changes back to data.yaml
    with open(target_yaml, 'w') as f:
        yaml.dump(data, f, sort_keys=False)

    global data_yaml_path
    data_yaml_path = target_yaml

    print(f"\n🎉 SUCCESS: The paths are locked, and the AI is ready to learn Swahili!")
else:
    print("⚠️ Error: Dataset folder not found. Please ensure the Roboflow download was successful.")
````

### 🗣️ Phase 3-A: The Data Auditor (Visual Confirmation)
As a Project Manager, you will often face **"Dirty Data."** When we inspected **ID 4 (Unknown_4),** we found a mix of everything: Minibuses, distant people, and blurry cars. If we teach the AI that this "mess" is a specific category, the AI will get confused and its "IQ" (Accuracy) will drop.

**The PM's Strategy:**
We will label ID 4 as **"Misc (Ignore)"**. This tells the AI: "Look at this, but don't try too hard to memorize it." This protects the learning quality of our important classes like **Pikipiki** and **Bajaj**.

````python
import os
import yaml
import glob
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Retrieve class names from the kernel state's 'data' variable
# Make sure 'data' is accessible and has 'names' key
if 'data' in locals() and 'names' in data:
    class_names = data['names']
    num_classes = len(class_names)
    print(f"Detected {num_classes} class IDs with remapped names: {class_names}")

    # Define paths
    # Using 'data["train"]' which is already absolute and points to images
    train_images_path = data['train'] # e.g., '/content/dhaka-1/train/images'
    train_labels_path = os.path.join(os.path.dirname(train_images_path), 'labels') # Correctly derive labels path

    if not os.path.exists(train_images_path) or not os.path.exists(train_labels_path):
        print(f"Error: Training image path '{train_images_path}' or label path '{train_labels_path}' does not exist.")
    else:
        # Dictionary to store a sample image path for each *remapped* class
        sample_images = {}

        # Iterate through label files to find a sample for each class
        label_files = glob.glob(os.path.join(train_labels_path, '*.txt'))

        # Get the number of occurrences of each class (using remapped names)
        unique_class_names = sorted(list(set(class_names))) # Get unique class names for counts
        class_counts = {name: 0 for name in unique_class_names}

        print("Scanning label files for class samples and counts...")
        for label_file in label_files:
            with open(label_file, 'r') as f:
                lines = f.readlines()
                for line in lines:
                    parts = line.strip().split()
                    if parts:
                        class_id = int(parts[0]) # This is the original class ID from the label file
                        if 0 <= class_id < num_classes:
                            remapped_class_name = class_names[class_id] # This is the remapped name

                            class_counts[remapped_class_name] += 1

                            # Store sample image if not already stored for this *remapped* class
                            if remapped_class_name not in sample_images:
                                image_basename = os.path.basename(label_file).replace('.txt', '')
                                found_image_path = None
                                for ext in ['.jpg', '.jpeg', '.png']:
                                    candidate_image_path = os.path.join(train_images_path, image_basename + ext)
                                    if os.path.exists(candidate_image_path):
                                        found_image_path = candidate_image_path
                                        break

                                if found_image_path:
                                    sample_images[remapped_class_name] = {'image_path': found_image_path, 'label_path': label_file}

        print("\nClass Counts in Training Data:")
        for name, count in class_counts.items():
            print(f"- {name}: {count} instances")

        if not sample_images:
            print("No sample images found for any class with ground truth labels.")
        else:
            print("\nDisplaying one sample image per detected class (with ground truth labels):")

            # Define a default font if ImageFont.truetype fails
            try:
                font_path = "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf" # Common Colab font
                if not os.path.exists(font_path):
                    font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf" # Another common Colab font
                font = ImageFont.truetype(font_path, 15)
            except IOError:
                font = ImageFont.load_default()
                print("Warning: Custom font not found, using default font for labels.")

            # Sort classes for consistent display
            sorted_sample_images = sorted(sample_images.items())

            fig, axes = plt.subplots(len(sorted_sample_images), 1, figsize=(10, 5 * len(sorted_sample_images)))
            if len(sorted_sample_images) == 1: # Handle case with a single subplot
                axes = [axes]

            for i, (class_name_key, paths) in enumerate(sorted_sample_images):
                image_path = paths['image_path']
                label_path = paths['label_path']

                img = Image.open(image_path).convert("RGB")
                draw = ImageDraw.Draw(img)
                img_width, img_height = img.size

                # Load and draw bounding boxes
                with open(label_path, 'r') as f:
                    for line in f.readlines():
                        parts = line.strip().split()
                        if parts:
                            class_id = int(parts[0]) # Original class ID
                            # Use remapped_class_name for display, check if it matches the current sample key
                            current_label_name = class_names[class_id]

                            # Only draw bounding box if the object belongs to the class we are showcasing for this subplot
                            # This ensures each subplot clearly represents its target class, even if other objects are present.
                            # To show all objects, remove this 'if current_label_name == class_name_key:' condition.
                            # For simplicity and to clearly show 'sample for Class X', we draw only for that class.
                            # Let's actually draw all for context, and label the image with the primary class it represents

                            x_center, y_center, width, height = map(float, parts[1:])

                            # Convert YOLO format (normalized center, width, height) to pixel coordinates (xmin, ymin, xmax, ymax)
                            xmin = int((x_center - width / 2) * img_width)
                            ymin = int((y_center - height / 2) * img_height)
                            xmax = int((x_center + width / 2) * img_width)
                            ymax = int((y_center + height / 2) * img_height)

                            # Draw rectangle
                            draw.rectangle([xmin, ymin, xmax, ymax], outline="red", width=2)

                            # Draw class name
                            label_text = class_names[class_id]
                            draw.text((xmin, ymin - 15), label_text, fill="red", font=font)

                axes[i].imshow(img)
                axes[i].set_title(f"Sample Image (Ground Truth) for Class: {class_name_key}")
                axes[i].axis('off')

            plt.tight_layout()
            plt.show()

else:
    print("Error: 'data' variable or 'names' key not found in kernel state. Please ensure data.yaml was processed and 'data' variable is correctly populated.")
````

````python
import os
import cv2
import random
import shutil

# 1. Define Paths 📍
img_dir = '/content/dhaka-1/train/images'
lbl_dir = '/content/dhaka-1/train/labels'
save_dir = '/content/check_samples'

# Clear existing folder for a fresh inspection 🧹
if os.path.exists(save_dir):
    shutil.rmtree(save_dir)
os.makedirs(save_dir, exist_ok=True)

target_id = "4"

# 2. Search for label files containing ID 4 🔍
all_labels = [f for f in os.listdir(lbl_dir) if f.endswith('.txt')]
target_list = []
for lbl in all_labels:
    with open(os.path.join(lbl_dir, lbl), 'r') as f:
        if any(line.split()[0] == target_id for line in f):
            target_list.append(lbl)

# 3. Extract 4 random samples and save them individually 🎲
if len(target_list) >= 4:
    samples = random.sample(target_list, 4)
    print(f"✅ System: Found {len(target_list)} images with ID 4. Extracting 4 random samples for PM review.")

    for idx, lbl_file in enumerate(samples):
        img_name = lbl_file.replace('.txt', '.jpg')
        img_path = os.path.join(img_dir, img_name)
        img = cv2.imread(img_path)

        if img is not None:
            h, w, _ = img.shape
            with open(os.path.join(lbl_dir, lbl_file), 'r') as f:
                for line in f:
                    parts = line.split()
                    cls, x, y, nw, nh = map(float, parts)
                    if int(cls) == int(target_id):
                        # Draw bounding box 🟥
                        x1, y1 = int((x - nw/2) * w), int((y - nh/2) * h)
                        x2, y2 = int((x + nw/2) * w), int((y + nh/2) * h)
                        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 3)
                        cv2.putText(img, "Suspicious ID 4", (x1, y1-10),
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

            # Save the individual file 💾
            save_path = f"{save_dir}/sample_check_{idx+1}.jpg"
            cv2.imwrite(save_path, img)
            print(f"📥 Saved: {save_path}")

    print("\n[👨‍💻 PM Action Required: How to check the samples]")
    print("1. Click the 'Folder' icon 📁 on the left sidebar in Colab.")
    print("2. Open the 'check_samples' folder.")
    print("3. Double-click the images to view them. 🖼️")
else:
    print("⚠️ Warning: Not enough images found for this ID.")
````
